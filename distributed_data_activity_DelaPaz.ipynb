### `PART 2` BIG DATA PROCESSING

```python
# Load dataset

from pyspark.sql.functions import col

df = spark.read.csv("phivolcs_earthquake_data.csv", header=True, inferSchema=True)
df.show(5)
```

*   Is the data processed sequentially or distributed?

    **The data is processed in a distributed way. From what I know, Spark divides the data in smaller portions and processes them across multiple cores.**

*   What does Spark do internally?

    **Spark breaks data into smaller portions, and distributes it to multiple processing units or cores. Afterwards, the operations it starts to do is parallel.**



---

### Filter Data

```python
filtered_df = df.filter(col("Magnitude") > 3)
filtered_df.show()
```

### Group and Aggregate

```python
grouped_df = df.groupBy("Specific_Location").count()
grouped_df.show()
```

### Sort Results

```python
sorted_df = grouped_df.orderBy("count", ascending=False)
sorted_df.show()
```
